```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib
matplotlib.rc('xtick',labelsize=30)
matplotlib.rc('ytick',labelsize=30)

import seaborn as sns
sns.set(style='whitegrid',color_codes=True)
from statsmodels.tsa.stattools import acf,pacf
```

```python
axline = np.linspace(0,10,500)

normal_nums = np.random.normal(0,5,500) #normally distributed
```

```python
expon_data = np.exp(axline**0.5)

y = normal_nums + expon_data
plt.figure(figsize=(20,10))
plt.plot(axline,y)
```

```python
# auto correlation function
ACF_lag = acf(y,nlags=200)

plt.figure(figsize=(20,10))
plt.plot(ACF_lag,marker='o')
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(y)),linestyle='--',color='blue')
plt.axhline(y=1.96/np.sqrt(len(y)),linestyle='--',color='blue')
plt.title('Auto Correlation Function')
plt.xlabel('Lag')
plt.ylabel('ACF')
plt.tight_layout()
plt.show()
```

Your time series has strong autocorrelation, especially at small lags.

The slow decay of ACF suggests that the series is non-stationary



```python
# partial auto correlation function
PACF_lag = pacf(y,nlags=30,method='ols')

plt.figure(figsize=(20,10))
plt.plot(PACF_lag,marker='o')
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(y)),linestyle='--',color='blue')
plt.axhline(y=1.96/np.sqrt(len(y)),linestyle='--',color='blue')
plt.title('Partial Auto Correlation Function')
plt.xlabel('Lag')
plt.ylabel('PACF')
plt.tight_layout()
plt.show()
```

So for prediction we should use auto regression model with atmost 7 lags
